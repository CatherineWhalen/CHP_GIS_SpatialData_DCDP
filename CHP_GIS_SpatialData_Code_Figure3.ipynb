```python
import pandas as pd
# create dataframe from Coding Template
coding_template_df = pd.read_csv('/content/IS262A_Coding_Template.csv')
# separate countries by ;
countries_series = coding_template_df["Country of Study"].str.split(";").explode().str.strip()
# count occurrences without duplicates
countries_count_df = (countries_series.value_counts().reset_index())
# create column headers 'NAME_ENGLI' in accordance with GADM country field
countries_count_df.columns = ['NAME_ENGLI', 'count']
# make 'count' field numeric
countries_count_df['count'] = pd.to_numeric(countries_count_df['count'])
# save as csv
countries_count_df.to_csv('IS262A_Coding_Template_countries_list.csv', index=False)
```
